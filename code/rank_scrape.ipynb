```python
import requests
from bs4 import BeautifulSoup
```

```python
html = requests.get('https://www.4icu.org/us/').text
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup 
```

```python
table = soup.find('table', class_ = 'table table-hover')
table
```

```python
# Initialize empty lists
numbers = []
hrefs = []
names = []
locations = []

# Extract data
for tr in table.find_all('tr'):
    # Extract the number inside <b> tag
    b_tag = tr.find('b')
    if b_tag:
        number = b_tag.text
        numbers.append(number)
    else:
        numbers.append(None)
    
    # Extract the href and the name of the institution
    a_tag = tr.find('a')
    if a_tag:
        hrefs.append(a_tag['href'])
        names.append(a_tag.text)
    else:
        hrefs.append(None)
        names.append(None)
    
    # Extract the location
    td_tags = tr.find_all('td')
    if td_tags:
        location = td_tags[-1].text
        locations.append(location)
    else:
        locations.append(None)
```

```python
print("Numbers:", numbers)
print("Hrefs:", hrefs)
print("Names:", names)
print("Locations:", locations)
```

```python
import pandas as pd
df = pd.DataFrame({
    'Number': numbers[1:-1],
    'Href': hrefs[1:-1],
    'Name': names[1:-1],
    'Location': locations[1:-1]
})
df
```

```python
df.to_csv('../data/4icu_rankings.csv', index=False)
```
